```python
import numpy as np
import matplotlib.pyplot as plt

# Function to minimize
def function(x):
    return (x + 3) ** 2

# Derivative of the function
def derivative(x):
    return 2 * (x + 3)

# Gradient Descent Algorithm
def gradient_descent(x, learning_rate, epochs):
    x_values = []
    y_values = []
    for i in range(epochs):
        gradient = derivative(x)
        x = x - learning_rate * gradient
        x_values.append(x)
        y_values.append(function(x))
    return x_values, y_values

# Function to find local minima
def find_local_minima(x_values, y_values):
    local_minima_index = y_values.index(min(y_values))
    return x_values[local_minima_index], y_values[local_minima_index]

# Initial point and hyperparameters
starting_point = 2
learning_rate = 0.1
epochs = 20

# Perform Gradient Descent
x_vals, y_vals = gradient_descent(starting_point, learning_rate, epochs)

# Find local minima
minima_x, minima_y = find_local_minima(x_vals, y_vals)

# Print the local minimum value
print(f'Local minimum occurs at x = {minima_x}, y = {minima_y}')

# Plotting the function
x = np.linspace(-10, 4, 1000)
y = function(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y=(x+3)^2')
plt.scatter(x_vals, y_vals, color='red', label='Gradient Descent path')
plt.scatter(minima_x, minima_y, color='green', marker='o', s=100, label='Local Minima')
plt.title('Gradient Descent to Find Local Minimum')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
